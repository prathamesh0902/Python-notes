## Variables

- No need of explicit delcaration
- Variable name starts with alphabet or underscore
- Variable name contains alphanumeric + underscore
- Variables are case sensitive

```python
# Not a variable name
# 2num, @num, df2.5, student-id

# Special variables with double underscores in prefix and suffix
# __varname__
```

## Data Type

> It represents which type of value is stored in a variable

No need for explicit declaration. Implicit declaration

- None
- Numeric
- Sequence
- Map
- Set

```python
num1 = 88.6   # float
num2 = 88     # int
```

**None**
> A variable which has no value

```python
num = None
```

```python
type(num)
```

**Numeric**
> A variable which have numerical value, on which arithmatic operations can be performed
- `int` - that cannot be written in the form of p/q E.g.: 24, -67
- `float` - that can be written in the form of p/q E.g.: 24.66, -768.95
- `complex`- combination of real and imaginary numbers:  a + bj

```python
num3, num4, num5 = 28, -85.6, 7+3.4j
type(num3), type(num4), type(num5)
```

```python
comp1, comp2 = complex(5, 6), complex(4, 8)
comp1 + comp2
```

**Type Casting**
> Technique to convert data types of a variable with set of rules
- int <> float
- int / float > complex

```python
num6 = 99
num7 = float(num6)
num8 = 67.99
num9 = int(num8)
num10 = complex(num6)
num7, num9, num10, type(num7), type(num9), type(num10)
```

**Sequence**
> String is a sequence data type in which we can store combination of characters and any type of value within "" or ' '
- allows indexing
- allows slicing
- is immutable

```python
str1 = 'Dagger Knight 99'
type(str1)
```

```python
str1[0], str1[-1], str1[-4], str1[0:13:2]
```

```python
str1[-1:-8:-1], str1[-6:], str1[-6::-1], str1[-11:-6:], str1[::-1]
```

```python
# not immutable
# str1[0 = 'T']

str1 = str1.replace('g', 'G')
str1
```

**String Operations**

```python
str2 = str1.replace("DaGGer", "Panther")
str3 = str1.split('G')
str4 = "g".join(str3)

str2, str3, str4
```
