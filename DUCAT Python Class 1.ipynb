## Variables

**Naming comvention** : Set of rules of defining a variable
- No need of explicit delcaration
- Variable name starts with alphabet or underscore
- Variable name contains alphanumeric + underscore
- Variables are case sensitive

```python
# Not a variable name
# 2num, @num, df2.5, student-id

# Special variables with double underscores in prefix and suffix
# __varname__
```

**Number Systems**

```python
# Leading zeros in decimal integer literals are not permitted;
# 04, 068
```

```python
num11 = 0b1000
num12 = 0o2343
num13 = 0X1A
num11, num12, num13
```

```python
bin(9),oct(9),hex(9)
```

**Memory management**

```python
num14 = 89
num15 = 88
num14 = 99

## Heap memory

## size reduces   > store at same place, release the memory
## size increases > store at new place
```

```python
num16 = 98  + 1

## will have same memory location as num14
```

```python
id(num14), id(num16)
```

## Data Type

> It represents which type of value is stored in a variable

No need for explicit declaration. Implicit declaration

- None
- Numeric
- Sequence
- Map
- Set

```python
num1 = 88.6   # float
num2 = 88     # int
```

### None
> A variable which has no value

```python
num = None
```

```python
type(num)
```

### Numeric
> A variable which have numerical value, on which arithmatic operations can be performed
- `int` - that cannot be written in the form of p/q E.g.: 24, -67
- `float` - that can be written in the form of p/q E.g.: 24.66, -768.95
- `complex`- combination of real and imaginary numbers:  a + bj

```python
num3, num4, num5 = 28, -85.6, 7+3.4j
type(num3), type(num4), type(num5)
```

```python
comp1, comp2 = complex(5, 6), complex(4, 8)
comp1 + comp2
```

**Type Casting**
> Technique to convert data types of a variable with set of rules
- int <> float
- int / float > complex

```python
num6 = 99
num7 = float(num6)
num8 = 67.99
num9 = int(num8)
num10 = complex(num6)
num7, num9, num10, type(num7), type(num9), type(num10)
```

### Sequence
**String**

> String is a sequence data type in which we can store combination of characters and any type of value within "" or ' '
- allows indexing
- allows slicing
- is immutable

```python
str1 = 'Dagger Knight 99'
type(str1)
```

```python
str1[0], str1[-1], str1[-4], str1[0:13:2]
```

```python
str1[-1:-8:-1], str1[-6:], str1[-6::-1], str1[-11:-6:], str1[::-1]
```

```python
# not immutable
# str1[0] = 'T'

str1 = str1.replace('g', 'G')
str1
```

**String Operations**

```python
str2 = str1.replace("DaGGer", "Panther")
str3 = str1.split('G')
str4 = "g".join(str3)

str2, str3, str4
```

```python
str5 = "     Dagger   Knight  99      "
str5.lstrip(), str5.rstrip(), str5.strip(), len(str5), str5.count("g")

## Remove all Extra spaces
## How many number of extra spaces
```

**Bytes**
> It is a sequence data type in which we can store only `int` with fix range 0 to 255
- allows indexing
- allows slicing
- is immutable

```python
seq1 = bytes(12)
seq1
```

- used for encoding / decoding
- saving passwords
- Graph data structure

```python
seq2 = bytes([3, 6, 2, 5, 7, 2, 1, 0, 2, 4])
seq2
```

```python
seq2[2], seq2[-3], seq2[::-1], list(seq2[::-1])
```

```python
len(seq2), seq2.count(2), seq2.index(2)
```

**Bytes Array**
> Silimar to bytes data type, difference only being that it is mutable
- allows indexing, slicing, mutable

```python
seq3 = bytearray([3, 6, 4, 6, 2, 4, 1, 0, 2, 4])
seq3
```

```python
seq3[2], seq3[-3], seq3[::-1], list(seq3[::-1])
```

```python
seq3[2] = 8
list(seq3)
```

**Lists**
> Sequence data type in which we can store any value
- allows indexing, slicing, mutable

```python
seq4 = []
type(seq4)
```

```python
seq4 = [32, 67.9, "Dagger", "Knight 99", 5 + 6j]
seq4[0], seq4[-1], seq4[::-1]
```

```python
seq4[2] = "Panther"
seq4
```

```python
seq4.append(4), seq4.insert(1,8), seq4.append([10, 20]), seq4.extend([100, 200])
seq4
```

```python
seq4.pop(), seq4.pop(2), seq4.remove(4) # seq.clear()
seq4
```

```python
len(seq4), seq4.count(8), seq4.index(8)
```

```python
seq5 = [23, 53, 32, 78, 32, 12, 11]
seq5.sort()
seq5, seq5[::-1]
```

```python
min(seq5), max(seq5), sum(seq5)
```

```python
seq4.extend(seq5)
seq4, seq5 + seq5, 2 * seq5
# 20 + seq5 not supported
```

```python
seq6 = seq5     # copying at the same address
seq7 = seq5[:]  # cloning the same data
seq6, seq7
```

**Range**
> It is a sequence data type in which only `int` values are stored with a fixed defined range
- indexing, slicing, not mutable

```python
seq12 = range(10, 51, 2)          # range(10, 40.5, 1) does not take float
type(seq12), seq12, list(seq12), list(range(6)), list(range(6,0,-1))
```

```python
# not immutable
# range[0] = 99

# Applications
# List of odd numbers               range(10,51,2)
# List of even numbers              range(11,51,2)
# Multipliaction Tables             range(10,101,2)
```

- range > list

```python
seq13 = list(seq12)
# seq13 = range(seq13)   not convertible
```
