```python
class detect:
    def evenodd(self, n1):
        if (n1 % 2 == 0):
            print(n1, "is even")
        else:
            print(n1, "is odd")

    def polydrom(self, n):
        # Converting to string to check for palindrome
        if str(n) == str(n)[::-1]:
            print(n, "is a Palindromic Number")
        else:
            print(n, "is not a Palindromic Number")
```

```python
dt1 = detect()
dt1.evenodd(99)
```

```python
dt1.polydrom(99)
```

```python
class emp:
    def __init__(self):               # special method for initialization: constructor
        self.nm = 'Raghav'
        self.salary = 45600
        print("Employee created")     # constructors are called by default

    def show(self): # also an instance method
        print(f"\tName: {self.nm} \n\tSalary: {self.salary}")

    # @classsmethod
    # @staticmethod
```

```python
emp1 = emp()
emp2 = emp()
emp1.show()
```

```python
emp1.nm, emp1.salary      # calling instance variable directly
```

```python
emp2.nm = 'Rahul'            # update variable values
emp2.salary = 50000
emp2.show()
```

```python
class create_emp:
    cmp_name = 'DUCAT'                     # class variable
    
    def __init__(self, nm_ip, salary_ip):  # special method for initialization: constructor
        self.nm = nm_ip
        self.salary = salary_ip
        print("Employee created")          # constructors are called by default

    def get_show(self):                    # Accessor instance method
        print(f"\tClass: {create_emp.cmp_name} \n\tName: {self.nm} \n\tSalary: {self.salary}")

    def set_increment(self):               # Mutate instance method
        self.salary = self.salary + self.salary * eval(input("% increment :"))

    @classmethod                           # to work on class variables
    def modify_cls(cls):
        cls.cmp_name = input("Enter updated name :")

    @staticmethod
    def calc(n1, n2):                      # local variables not accessible outside class
        print(f"Addition: {n1 + n2}")
```

```python
emp3 = create_emp("Andy", 250000)
emp3.get_show()
```

```python
emp3.set_increment()
emp3.get_show()
```

```python
create_emp.cmp_name = "DUCAT India"       # update class variable name
emp3.get_show()
```

```python
emp3.modify_cls()
emp3.get_show()
```

```python
emp3.calc(40, 50)
```

# Inheritance
- Single inheritance
- Multiple inheritance

```python
class Mother():
    def __init__(self):
        self.propertyM = 320000

    def get_show(self):
        print(f"Self Property = {self.propertyM}")

class Son(Mother):
    def __init__(self):
        self.propertyS = 32000
        
    def get_show(self):
        print(f"Self Property = {self.propertyS}")
    #     super().get_show()                  # not accessible due to method overriding
    
    # def get_show_parent(self):
    #     print(f"Parent Property = {self.propertyM}")

class Daughter(Mother):
    def __init__(self):
        self.propertyD = 3200
        super().__init__()                    # avoid method overriding

    def get_show(self):
        print(f"Self Property = {self.propertyD}")

    def get_show_parent(self):
        print(f"Parent Property = {self.propertyM}")

    def get_show_heir(self):
        print(f"Heirarchy Property = {self.propertyD}")
        super().get_show()                   # avoid method overriding

    def total(self):
        print(f"Total Property = {self.propertyD + self.propertyM}")
```

```python
son1 = Son()
son1.get_show()
```

```python
# son1.get_show_parent()      # not accesible due to method overriding
```

```python
dt1 = Daughter()               # method overriding
dt1.get_show()                 # __init__ class properties overridden
```

```python
dt1.get_show_parent()
```

```python
dt1.get_show_heir()
```

```python
class Father:
    def __init__(self):
        self.propertyF = 420000
        
    def get_show(self):
        print(f"Property = {self.propertyF}")
```

```python
class Son2(Mother, Father):
    pass

son2 = Son2()
son2.get_show()
```

```python
class Son3(Father, Mother):
    pass

son3 = Son3()
son3.get_show()
```

```python
class Step_Mother():
    def __init__(self):
        self.propertySM = 520000
        super().__init__()

    def get_show(self):
        print(f"Property = {self.propertySM}")
        super().get_show()
```

```python
class ide():
    def compile(self):
        print("Compiled")
    def run(self):
        print("Ran")
    def debug(self):
        print("Debugged")

class coder():
    def show(self, other):
        other.compile()
        other.run()

ide1 = ide()
vscode = coder()
vscode.show(ide1)
```

## 2. Method overloading

```python
class Calculator:
    """
    Demonstrates different types of method overloading in Python
    """

    # 1️⃣ Overloading using Default Arguments
    def add_default(self, a, b=0, c=0):
        return a + b + c

    # 2️⃣ Overloading using *args (Variable Length Arguments)
    def add_args(self, *numbers):
        return sum(numbers)

    # 3️⃣ Overloading using Type Checking
    def add_type(self, a, b):
        if isinstance(a, int) and isinstance(b, int):
            return a + b
        elif isinstance(a, str) and isinstance(b, str):
            return a + " " + b
        else:
            return "Unsupported types"

    # 4️⃣ Overloading using Keyword Arguments (**kwargs)
    def add_kwargs(self, **kwargs):
        return sum(kwargs.values())

    # 5️⃣ Overloading using Conditional Argument Count
    def add_conditional(self, *args):
        if len(args) == 2:
            return args[0] + args[1]
        elif len(args) == 3:
            return args[0] + args[1] + args[2]
        else:
            return "Invalid number of arguments"


# ------------------ USAGE ------------------

calc = Calculator()

print(calc.add_default(5))
print(calc.add_default(5, 10))
print(calc.add_default(5, 10, 20))

print(calc.add_args(1, 2, 3, 4))

print(calc.add_type(5, 10))
print(calc.add_type("Hello", "World"))

print(calc.add_kwargs(a=10, b=20, c=30))

print(calc.add_conditional(10, 20))
print(calc.add_conditional(10, 20, 30))
```

## 4. Operator Overloading

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overload + operator
    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def show(self):
        print(f"({self.x}, {self.y})")


# -------- USAGE --------
p1 = Point(2, 3)
p2 = Point(4, 5)

p3 = p1 + p2   # calls __add__()
p3.show()
```
