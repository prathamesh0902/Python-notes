## Arguments

```python
def student_profile(name, age, country="USA"):
    """
    Demonstrates Python argument types.
    """
    print(f"Name: {name} (Positional)")
    print(f"Age: {age} (Keyword/Positional)")
    print(f"Country: {country} (Default)")

def student_activity(*grades, **extra_curr):
    # *grades is Variable Length (Tuple)
    if grades:
        avg = sum(grades) / len(grades)
        print(f"Average Grade: {avg:.2f} (Variable Length)")
        
    # **extra_info is Variable Keyword (Dictionary)
    for key, value in extra_curr.items():
        print(f"{key.capitalize()}: {value} (Variable Keyword)")

        
# 1. Positional Arguments: Passed in order ('Alice' -> name, 24 -> age)
# 2. Keyword Arguments: Explicitly named (age=24)
# 3. Default Arguments: 'country' is used if not provided
# 4. Variable Length (*args): 88, 92, 95 are packed into a tuple
# 5. Variable Keyword (**kwargs): hobby and major are packed into a dict
```

```python
student_profile(
    "Alice",                  # Positional
     age=24,                  # Keyword
     country = "Canada",      # Overriding Default
)
```

```python
student_activity(
    88,92,76,52,
    hobby = "Chess",
    major = "AI Engineering"
)
```

## Anonymous function
> Using `lambda` keyword
- does not work for loop
- does not work for elif

```python
sum = lambda x,y: x + y
sum(20, 50)
```

```python
addsubmul = lambda n1, n2 : (n1 + n2, n1 - n2, n1 * n2)
addsubmul(20, 30)
```

```python
detect_even = lambda num: 'Even' if num %2 == 0 else 'Odd'
detect_even(20), detect_even(21)
```

```python
category = lambda num: "A" if num < 20 else "B" if num < 40 else "C"
category(37), category(10), category(51)
```

> `filter`, `map` and `reduce`

```python
list4 = [12,34,54,234,55, 45, 134,32,67,88,98,987]
list(filter(lambda n: n % 2 == 0, list4))
```

```python
list(filter(lambda n: n % 2 == 0 and n == int(str(n)[::-1]), list4))
```

```python
list(map(lambda n: n // 2 if n % 2 == 0 else n // 3, list4))
```

```python
list(map(lambda n: "A" if n < 100 else "B" if n < 200 else "C", list4))
```

```python
from functools import reduce

reduce(lambda n1, n2: n1 + n2 / 2, list4), sum(list4)
```

```python
class detect:
    def evenodd(self, n1):
        if (n1 % 2 == 0):
            print(n1, "is even")
        else:
            print(n1, "is odd")

    def polydrom(self, n):
        # Converting to string to check for palindrome
        if str(n) == str(n)[::-1]:
            print(n, "is a Palindromic Number")
        else:
            print(n, "is not a Palindromic Number")
```

```python
dt1 = detect()
dt1.evenodd(99)
```

```python
dt1.polydrom(99)
```
